## 1. Herramientas de Inteligencia Artificial utilizadas

Durante el desarrollo del proyecto se utilizaron principalmente **ChatGPT** y **Claude (Anthropic)** como herramientas de apoyo al aprendizaje, resolución de dudas y mejora del proceso de desarrollo.

* **ChatGPT:** se utilizó para aclarar conceptos relacionados con el preprocesamiento de datos, explicar el funcionamiento de fragmentos de código, interpretar métricas de evaluación, sugerir diferentes formas de implementar ciertas tareas y ayudar a formular prompts más específicos para obtener mejores respuestas.

* **Claude:** se utilizó como un asistente tipo tutor para revisar el código desarrollado, responder preguntas sobre errores específicos, proponer alternativas de implementación, orientar el diseño de los modelos de redes neuronales artificiales (ANN) y apoyar la integración de componentes como la API y el dashboard.

En ambos casos, las herramientas fueron utilizadas como apoyo para comprender el proceso y tomar decisiones informadas, mientras que el desarrollo, adaptación, ejecución y validación del código final fueron realizados por el equipo.

## 2. Tareas específicas en las que se utilizó la IA

Las herramientas de IA fueron utilizadas en diferentes etapas del proyecto, entre ellas:

* Explicación de conceptos relacionados con el preprocesamiento y análisis exploratorio de datos.
* Interpretación de métricas de evaluación como accuracy, precision, recall, F1-score, MAE y RMSE.
* Resolución de errores de programación y depuración del código.
* Orientación sobre la codificación de variables categóricas y limpieza de datos.
* Recomendaciones para el diseño y ajuste de las arquitecturas de los modelos ANN.
* Sugerencias para mejorar hiperparámetros como learning rate, número de capas, batch size y uso de class_weight.
* Explicación sobre buenas prácticas, como aplicar el escalado después del train_test_split.
* Apoyo en la integración del API desarrollada con FastAPI.
* Ayuda para resolver inconvenientes relacionados con versiones de Keras y TensorFlow al cargar modelos.
* Sugerencias para mejorar la visualización y presentación del dashboard desarrollado en Streamlit.

## 4. Reflexión sobre el aprendizaje

El uso de herramientas de inteligencia artificial fue de gran ayuda para reforzar y complementar el aprendizaje durante el desarrollo del proyecto. En varias ocasiones sirvieron para aclarar dudas sobre conceptos vistos en clase que no recordaba completamente o sobre cuya aplicación práctica existía inseguridad. Más que proporcionar respuestas inmediatas, las explicaciones permitieron comprender por qué una solución era adecuada y cuáles eran sus ventajas, limitaciones y posibles alternativas.

Además, la IA funcionó como una herramienta de retroalimentación constante. Permitió cuestionar decisiones, comparar diferentes enfoques y reflexionar sobre cuál era la alternativa más apropiada según el contexto del problema. También ayudó a comprender errores que habrían sido difíciles de resolver sin orientación, como los relacionados con versiones de TensorFlow y Keras o la integración de distintos componentes del sistema.

Una de las principales lecciones aprendidas fue que las sugerencias generadas por la IA no siempre funcionan correctamente en el primer intento. Fue necesario probar, ajustar, validar resultados y realizar modificaciones hasta obtener modelos satisfactorios. Este proceso fortaleció el razonamiento crítico, la interpretación de resultados y la confianza para trabajar con técnicas de Machine Learning y redes neuronales.

## 5. Modificaciones realizadas al código sugerido por IA

Las recomendaciones proporcionadas por las herramientas de IA fueron revisadas, adaptadas y modificadas antes de incorporarlas al proyecto. Entre las principales modificaciones realizadas se encuentran:

* Adaptación del código para reemplazar valores específicos del dataset según las necesidades del proyecto.
* Conversión de la variable `TotalCharges` de texto a formato numérico mediante `pd.to_numeric()`.
* Ajuste del proceso de codificación de variables categóricas utilizando LabelEncoder únicamente en las variables correspondientes.
* Integración de las recomendaciones dentro de la estructura y estilo de programación trabajados en clase.
* Ajuste de arquitecturas, hiperparámetros y procesos de entrenamiento de las redes neuronales según los resultados obtenidos durante las pruebas.
* Modificación de rutas de archivos para adecuarlas al entorno local de desarrollo.
* Adaptación del código para utilizar los archivos generados por el equipo, como `customer_clean.csv`.
* Verificación de que la codificación utilizada en el dashboard coincidiera con la aplicada durante el preprocesamiento.
* Implementación de alternativas para guardar y cargar modelos cuando surgieron incompatibilidades entre versiones de librerías.

Las herramientas de IA fueron utilizadas como un apoyo para el aprendizaje y la resolución de dudas específicas; sin embargo, el análisis, la toma de decisiones, la validación de resultados y la implementación final del proyecto fueron responsabilidad del equipo.
