comparacion-Modelos-ANN

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Resultados obtenidos

```python
resultados = {
    'Modelo': ['Clasificación Binaria (ANN)', 'Regresión (ANN)'],
    'Métrica Principal': ['Accuracy', 'MAE'],
    'Valor': [0.84, 0.20],
    'Objetivo Mínimo': [0.84, 0.22],
    'Cumple': [' Sí', ' Sí']
}

df_resultados = pd.DataFrame(resultados)
print(df_resultados.to_string(index=False))
```

Gráfica comparativa

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(['Clasificación'], [0.84], color='#4A90D9')
axes[0].axhline(y=0.84, color='red', linestyle='--', label='Mínimo requerido')
axes[0].set_title('Modelo 1 - Accuracy')
axes[0].set_ylim(0, 1)
axes[0].legend()

axes[1].bar(['Regresión'], [0.20], color='#2ecc71')
axes[1].axhline(y=0.22, color='red', linestyle='--', label='Máximo permitido')
axes[1].set_title('Modelo 2 - MAE')
axes[1].set_ylim(0, 0.5)
axes[1].legend()

plt.suptitle('Comparación de Modelos ANN - Telco Churn')
plt.tight_layout()
plt.show()
```

Conclusión

```python
print("""
CONCLUSIÓN:
- Modelo 1 (Clasificación): Accuracy 0.84  cumple el mínimo requerido de 0.84
- Modelo 2 (Regresión): MAE 0.20  mejor que el mínimo requerido de 0.22
- Ambos modelos fueron entrenados con SMOTE para manejo de clases desbalanceadas
- Se recomienda usar el Modelo 1 para decisiones binarias (retener/no retener)
  y el Modelo 2 para priorizar clientes por nivel de riesgo (Bajo/Medio/Alto)
""")
```
